```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("../Dataset.csv")
dataset.head(5)
```

```python
dataset.columns
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataset["Longitude"], dataset["Latitude"], alpha=0.5)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Restaurant Distribution")
plt.show()
```

```python
group = dataset.groupby('City')
count = group.size()
```

```python
count.plot(kind='bar', figsize=(10, 6))
plt.title("No of Restaurants")
plt.xlabel("City")
plt.ylabel("No of Restaurants")
plt.show()
```

```python
dataset["City"].unique()
```

```python
len(dataset["City"].unique())
```

```python
df = dataset.copy()
```

```python
city_count = dict(df["City"].value_counts().sort_values(ascending=False)[:20])
city_count
```

```python
city = list(city_count.keys())
count= list(city_count.values())
y = np.arange(len(city))
plt.figure(figsize=(15, 5))
plt.bar(y, count, align='center', alpha=0.5)
plt.xticks(y, city, rotation=90)
plt.xlabel("City")
plt.ylabel("No of Restaurants")
plt.title("Top 20 Cities with Most Restaurants")
plt.show()
```

```python
locality_count = dict(df['Address'].value_counts().sort_values(ascending=False)[:20])
print(locality_count)
```

```python
country_count = dict(df['Country Code'].value_counts().sort_values(ascending=False)[:20])
country_count
```

```python
city = list(country_count.keys())
count= list(country_count.values())
y= np.arange(len(city))
plt.figure(figsize=(15, 5))
plt.bar(y, count, align='center', alpha=0.5)
plt.xticks(y, city, rotation=90)
plt.xlabel("Country Code")
plt.ylabel("No of Restaurants")
plt.title("Top 20 Countries with Most Restaurants")
plt.show()
```

```python
average= {}
for city, count in city_count.items():
  city_df = dataset[dataset['City'] == city]
  average_rating = city_df['Aggregate rating'].mean()
  average[city] = average_rating

sorted_average_ratings = {k: v for k, v in sorted(average.items(), key=lambda item: item[1], reverse=True)}
print("Average Ratings by City:")
for city, average_rating in sorted_average_ratings.items():
  print(city ,"  ",average_rating)
```

```python
city = list(sorted_average_ratings.keys())
average_ratings = list(sorted_average_ratings.values())
y = np.arange(len(city))
plt.figure(figsize=(15, 5))
plt.bar(y, average_ratings, align="center",color="green", alpha=0.5)
plt.xticks(y, city, rotation=90)
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.title("Average Rating of Restaurants by City")
plt.show()
```

```python
cuisines = dataset["Cuisines"].str.split(',').explode().value_counts().sort_values(ascending=False).to_dict()
cuisines
```

```python
average = {}
for city, count in city_count.items():
  city_df = dataset[dataset['City'] == city]
  average_price = city_df['Price range'].mean()
  average[city] = average_price
sorted_average_price = {k: v for k, v in sorted(average.items(), key=lambda item: item[1], reverse=True)}
print("Average Price by City:")
for city, average_price in sorted_average_price.items():
  print(city,"   ",average_price)
```

```python
city= list(sorted_average_price.keys())
average_price = list(sorted_average_price.values())
plt.figure(figsize=(15, 5))
plt.bar(city, average_price, color="green", width=0.5)
plt.xlabel("City")
plt.ylabel("Average Price")
plt.title("Average Price of Restaurants by City")
plt.xticks(rotation=90)
plt.show()
```
