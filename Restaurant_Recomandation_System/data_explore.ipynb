```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
data=dataset[["Longitude","Latitude"]]
data.head(4)
```

```python
distortions=[]
k=range(1,25)
for i in k:
    model=KMeans(n_clusters=i)
    model.fit(data)
    distortions.append(model.inertia_)
```

```python
fig,ax=plt.subplots(figsize=(12,6))
plt.plot(k,distortions,marker="o")
plt.xlabel("K")
plt.ylabel("Distortions")
plt.show()
```

```python
sil=[]
kmax=50
for k in range(2,kmax+1):
    model=KMeans(n_clusters=k).fit(data)
    labels=model.labels_
    sil.append(silhouette_score(data,labels,metric="euclidean"))
```

```python
sil
```

```python
model=KMeans(n_clusters=5,init="k-means++")
model.fit(data)
y=model.labels_
print("k=5","silhouette score:",silhouette_score(data,y,metric="euclidean"))
```

```python
topdata["cluster"]=model.predict(topdata[["Longitude","Latitude"]])
```

```python
def recomend_restaurants(df, longitude, latitude):
    location = pd.DataFrame([[longitude, latitude]], columns=["Longitude", "Latitude"])
    cluster = model.predict(location)[0]
    print("Predicted Cluster: ",cluster)
    return df[df["cluster"] == cluster].iloc[0:5][["Restaurant Name", "Longitude", "Latitude"]]
```
